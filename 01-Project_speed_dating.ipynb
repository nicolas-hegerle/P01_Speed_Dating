### <ins>Fuction definition (for later)</ins>

```python
# interesting to normalize because participants did not have the same scoring schem => scale 1 to 10 or distribute 100 points

def norm_attr_scores(col_names, df, inplace = False):
    
    """
    Normalizes the attribute scores by dividing each attr score by the sum of the attribute scores
    Ex: attr = 12, sinc = 10, intel = 8 will turn into
    attr = 12 / 30 = 0.4, sinc = 10 / 30 = 0.33, intel = 8 / 30 = 0.27
    
    Sum of the normalized attribute scores will be equal to one and the relative
    importance of each attr maintained for the particpant's ratings
    
    Arguments:
    col_names: list of column names you want to normalize. Must be given neighboring cols and given in order of increasing indice
    df: Pandas dataframe to which we want to apply the normalization

    Returns a python dictionary with keys = col_names, values = lists of all normalized attributes of length the df's number of rows.
    """

    # dict comprehension to initialize an empty dictionary with keys = col_names and values = empty lists
    normalized_attr = {col_name : [] for col_name in col_names} 

    # loop through the dataframe's index to get the sum of attribute scores
    for i in df.index:
        total_attr = df.loc[i, col_names].sum() 

    # loop throught the col_names to normalize the values per the total sum of all scores and store the 
        for j in col_names:            
            normalized_attr[j].append(df.loc[i,j]/total_attr)
    
    if inplace:
        normalized_attr = pd.DataFrame(normalized_attr, index = None)
        df[col_names] = normalized_attr[col_names]
        return df
    else:
        return normalized_attr
```

```python
# load the csv file into a Pandas dataframe
df = pd.read_csv('Speed Dating Data.csv', encoding="ISO-8859-1")

# display first five rows and the shape of the dataframe
print(f"\nThe dataframe has {df.shape[0]} rows and {df.shape[1]} columns\n")
print("First five rows of the dataframe:")
display(df.head())
```

```python
# How many unique participants?
print(f"There are {df['iid'].nunique()} unique participants throughout all the waves")

# Keep only unique participants
df_iid = df.drop_duplicates(subset = 'iid')
df_iid.reset_index(inplace = True, drop = True)
print(f"The dataset without duplicated iids has {df_iid.shape[0]} rows and {df_iid.shape[1]} columns")
# CAUTION: some columns cannot be used when duplicate iids are droped because they relate to scores given to partners during the evening

# calculate the proportion of men vs women...should be balanced but just to check
print(f"Women account for {round(df_iid['gender'].value_counts(normalize = True)[1]*100, 2)}% of the participants, men for {round(df_iid['gender'].value_counts(normalize = True)[0]*100, 2)}%")
```

```python
# build an interactive box plot showing the normalized scores per ethny

fig = go.Figure()
buttons = []
df_iid_melt.dropna(subset = ['ethnic_ori'], inplace=True)
ethnies = df_iid_melt['ethnic_ori'].unique()

for index, ethny in enumerate(ethnies):
    mask = df_iid_melt.loc[:, 'ethnic_ori'] == ethny
    df_graph = df_iid_melt.loc[mask]
    fig.add_trace(
        go.Box(
            x = df_graph['variable'], 
            y = df_graph['value'],
            visible = (True if index == 0 else False),
            name = ethny
        )
    )

    #Generate the buttons for the update menu
    vis = [False] + [False] * len(ethnies)
    vis[index] = True
    args = [{'visible' : vis}]
    buttons.append(go.layout.updatemenu.Button(label = ethnies[index], method = "update", args = args))

fig.add_trace(
    go.Box(
        x = df_iid_melt['variable'],
        y = df_iid_melt['value'],
        visible = (False),
        name = 'All'
    )
)

buttons.append(go.layout.updatemenu.Button(label = 'All', method = "update", args = [{'visible' : [False]*len(ethnies)+[True]}]))

fig.update_layout(
    title = {'text': 'Boxplot of normalized attribute scores', 'y':0.97, 'x':0.23, 'xanchor': 'left', 'yanchor': 'top'},
    updatemenus = [go.layout.Updatemenu(active = 0, buttons = buttons)], \
                autosize = False, \
                width = 900, \
                height = 600, \
                margin={'l':15, 'r':10, 'b':10, 't':40}
                )
```
